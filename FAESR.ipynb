```python
from gensim.models import Word2Vec


disease_lists = temp_df['pt'].apply(lambda x: [x]).tolist()
disease_model = Word2Vec(sentences=disease_lists, vector_size=100, window=5, min_count=1, workers=4)

disease_model.wv["Abdominal pain"]

prodai_lists = temp_df['prod_ai'].apply(lambda x: [x]).tolist()
prodai_model = Word2Vec(sentences=prodai_lists, vector_size=100, window=5, min_count=1, workers=4)
prodai_model.wv["CILTACABTAGENE AUTOLEUCEL"]
```

```python
disease_model.save('disease_model.bin')
prodai_model.save('prodai_model.bin')
```

```python
temp_df['pt']=[disease_model.wv[i] for i in temp_df['pt']]
temp_df['prod_ai']=[prodai_model.wv[i] for i in temp_df['prod_ai']]
```

```python



# Convert 'pt' and 'prod_ai' columns to DataFrames
pt_df = pd.DataFrame(temp_df['pt'].tolist(), index=temp_df.index)
prod_ai_df = pd.DataFrame(temp_df['prod_ai'].tolist(), index=temp_df.index)

# Rename columns to ensure uniqueness
pt_df.columns = [f'pt_{i}' for i in range(pt_df.shape[1])]
prod_ai_df.columns = [f'prod_ai_{i}' for i in range(prod_ai_df.shape[1])]

# Drop the original 'pt' and 'prod_ai' columns
temp_df = temp_df.drop(['pt', 'prod_ai'], axis=1)

# Concatenate the new columns with the existing DataFrame
temp_df = pd.concat([temp_df, pt_df, prod_ai_df], axis=1)

# Verify the final shape
print("Final temp_df shape:", temp_df.shape)  # Should be (number of rows, original columns + 200)
```

```python
x=temp_df.drop('outc_cod',axis=1)
y=temp_df['outc_cod']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
xtr,xt,ytr,yt=train_test_split(x,y)

```
